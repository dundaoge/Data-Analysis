```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
X=np.linspace(0,2*np.pi,100)
Y=np.sin(X)
Y1=np.cos(X)
```

```python
plt.title('Hello World!')
plt.plot(X,Y)
plt.plot(X,Y1)
```

# bar plot

```python
#vertical
data= [10,20,60,30]
plt.bar(range(len(data)),data)
```

```python
#horizontal
plt.barh(range(len(data)),data)
```

```python
data =[[10,20,23,50],
       [5,25,33,56],
      [32,45,66,18]]
X=np.arange(4)

plt.bar(X+0.00,data[0],color='r',width=0.25,label="A")
plt.bar(X+0.25,data[1],color='b',width=0.25,label="B")
plt.bar(X+0.50,data[2],color='g',width=0.25,label="C")

plt.legend() 
```

```python
data =[[10,20,23,50],
       [5,25,33,56],
      [32,45,66,18]]
X=np.arange(4)

plt.title('horizontal')
plt.bar(X,data[0],color='b',width=0.25)
plt.bar(X,data[1],color='r',width=0.25,bottom=data[0])#bottom表示horizontal底部数据
plt.bar(X,data[2],color='g',width=0.25,bottom=np.array(data[0])+np.array(data[1]))#需要转化为array数组才可以相加
```

# scatter points 
#衡量两个连续变量之间的相关性

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.randint(0,2,size=50)
area = np.pi * (20 * np.random.rand(N))**2 #调整大小

plt.scatter(x, y, c=colors, alpha=0.6, s=area)
```
